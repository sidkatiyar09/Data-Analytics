```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta

# Calculate start and end dates for the last 3 months
end_date = datetime.now()
start_date = end_date - timedelta(days=90)

tickers = ['AAPL', 'MSFT', 'NFLX', 'GOOG']

df_list = []

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    df_list.append(data)

df = pd.concat(df_list, keys=tickers, names=['Ticker', 'Date'])
print(df.head())
```

```python
import plotly.express as px

# Loop through each company and plot its moving averages and close prices
for ticker, group in df.groupby('Ticker_'):
    y_columns = []

    if ticker == 'MSFT':
        y_columns = ['MSFT_Close_MSFT', 'MA10', 'MA20']
    elif ticker == 'GOOG':
        y_columns = ['GOOG_Close_GOOG', 'MA10_GOOG', 'MA20_GOOG']
    elif ticker == 'NFLX':
        y_columns = ['NFLX_Close_NFLX', 'MA10_NFLX', 'MA20_NFLX']

    fig = px.line(group, x='Date_', y=y_columns, title=f"{ticker} Moving Averages")
    fig.show()
```

```python
import matplotlib.pyplot as plt

tickers = ['MSFT', 'GOOG', 'NFLX']
fig, axs = plt.subplots(len(tickers), 1, figsize=(12, 9), sharex=True)

for i, ticker in enumerate(tickers):
    group = df[df['Ticker_'] == ticker]
    ax = axs[i]

    if ticker == 'MSFT':
        ax.plot(group['Date_'], group['MSFT_Close_MSFT'], label='Close')
        ax.plot(group['Date_'], group['MA10'], label='MA10')
        ax.plot(group['Date_'], group['MA20'], label='MA20')
    elif ticker == 'GOOG':
        ax.plot(group['Date_'], group['GOOG_Close_GOOG'], label='Close')
        ax.plot(group['Date_'], group['MA10_GOOG'], label='MA10')
        ax.plot(group['Date_'], group['MA20_GOOG'], label='MA20')
    elif ticker == 'NFLX':
        ax.plot(group['Date_'], group['NFLX_Close_NFLX'], label='Close')
        ax.plot(group['Date_'], group['MA10_NFLX'], label='MA10')
        ax.plot(group['Date_'], group['MA20_NFLX'], label='MA20')

    ax.set_title(f"{ticker} Moving Averages")
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# List of unique tickers in your DataFrame
tickers = df['Ticker_'].unique()

# Create subplot grid: 2 rows x 2 cols for 4 tickers
fig = make_subplots(rows=2, cols=2, subplot_titles=[f"{ticker} Moving Averages" for ticker in tickers])

# Ticker-to-row-col mapping
positions = {(0): (1,1), (1): (1,2), (2): (2,1), (3): (2,2)}

for i, ticker in enumerate(tickers):
    group = df[df['Ticker_'] == ticker]
    row, col = positions[i]

    fig.add_trace(go.Scatter(x=group['Date_'], y=group['MSFT_Close_MSFT'], mode='lines', name=f'{ticker} Close'), row=row, col=col)
    fig.add_trace(go.Scatter(x=group['Date_'], y=group['MA10'], mode='lines', name=f'{ticker} MA10'), row=row, col=col)
    fig.add_trace(go.Scatter(x=group['Date_'], y=group['MA20'], mode='lines', name=f'{ticker} MA20'), row=row, col=col)
    fig.add_trace(go.Scatter(x=group['Date_'], y=group['NFLX_Close_NFLX'], mode='lines', name=f'{ticker} Close'), row=row, col=col)
    fig.add_trace(go.Scatter(x=group['Date_'], y=group['MA10_NFLX'], mode='lines', name=f'{ticker} MA10'), row=row, col=col)
    fig.add_trace(go.Scatter(x=group['Date_'], y=group['MA20_NFLX'], mode='lines', name=f'{ticker} MA20'), row=row, col=col)
    fig.add_trace(go.Scatter(x=group['Date_'], y=group['GOOG_Close_GOOG'], mode='lines', name=f'{ticker} Close'), row=row, col=col)
    fig.add_trace(go.Scatter(x=group['Date_'], y=group['MA10_GOOG'], mode='lines', name=f'{ticker} MA10'), row=row, col=col)
    fig.add_trace(go.Scatter(x=group['Date_'], y=group['MA20_GOOG'], mode='lines', name=f'{ticker} MA20'), row=row, col=col)
    # figplt, axs = plt.subplots(3, 1, figsize=(12, 6))

# Final layout
fig.update_layout(height=800, width=1200, title_text="Stock Moving Averages (All Tickers)", showlegend=False)

fig.show()
```
